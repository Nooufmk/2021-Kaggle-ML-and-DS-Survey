# Gender  

```python
plt.figure(figsize = (8,5))
labels= ['Man','Woman']
sns.set(style='darkgrid')
colors = ['lightskyblue', 'magenta']
gender = df['Q2'].value_counts()
plt.pie(gender,colors = colors, autopct='%1.1f%%', shadow=True,radius = 1.5,pctdistance = 1.2)
plt.legend(labels,loc = 'best')
plt.show()
```

## Let's see the questions answered by 50% of the Kaggle community 

```python
n = []
for i in df:
    if df[i].isnull().sum() > 25974/0.5:   #25974 in number of Respondents 
        n.append(i)
        
df1 = df.drop(n,axis = 1)
df1.drop(df1.index[0],inplace=True)
```

## Most used programming languages for 2021

```python
df7 = df[Q7]

df7.columns = ['Python', 'R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'MATLAB', 'None', 'Other']

plt.figure(figsize=(10,10))
df7.count().sort_values(ascending=True).plot(kind = 'pie');
plt.ylabel('Most used programming languages for 2021',fontsize = 20, weight = 'bold',color='green');
plt.show();
```

```python
plt.figure(figsize = (8,8))
df7.columns = ['Python', 'R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'MATLAB', 'None', 'Other']
df77 = df7.count()
sns.set(style='darkgrid')
plt.pie(df77, autopct='%1.1f%%', shadow=True,radius = 1.5,pctdistance = 1.2)
plt.legend( df7.columns ,loc = 'best')
plt.show()
df77
```

## Most used of (IDE's) in 2021

```python
Q9 = list(df2.loc[0][21:34])
df9 = df[Q9]

df9.columns = ['Jupyter (JupyterLab, Jupyter Notebooks, etc)',' RStudio','Visual Studio','Visual Studio Code (VSCode)',
              'PyCharm','Spyder','Notepad++','Sublime Text','Vim / Emacs','MATLAB','Jupyter Notebook','None','Other']

plt.figure(figsize=(10,10))
df9.count().sort_values(ascending=False).plot(kind = 'bar');
plt.ylabel("Most used (IDE's) for 2021",fontsize = 20, weight = 'bold',color='green');
plt.show();
```
